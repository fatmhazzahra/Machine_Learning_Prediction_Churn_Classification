##  1. Pendahuluan

### 1.1 Latar Belakang

Sesuai dengan perkembangan zaman, bisnis semakin modern dan tingkat kompetitif pun semakin tinggi. Hal ini menyebabkan kemampuan perusahaan untuk mempertahankan pelanggan (*customer retention*) menjadi aset yang jauh lebih berharga dibandingkan sekadar mengakuisisi pelanggan baru. Oleh karena itu, memahami pola perilaku pelanggan dan mendeteksi sinyal ketidakpuasan sejak dini bukan lagi sekadar opsi, melainkan kebutuhan strategis untuk keberlanjutan bisnis jangka panjang.

### 1.2 Rumusan Masalah Bisnis
Meskipun perusahaan memiliki data pelanggan yang melimpah, tidak semua perusahaan dapat memanfaatkannya dengan baik. Tantangan utama perusahaan adalah ketidakmampuan untuk mengidentifikasi pelanggan yang berisiko pergi (*churn*) sebelum mereka benar-benar berhenti berlangganan.

Masalah spesifik yang dihadapi meliputi:
* **Identifikasi Reaktif :** Perusahaan baru menyadari kehilangan pelanggan setelah pemutusan layanan terjadi, sehingga terlambat untuk melakukan tindakan pencegahan.
* **Revenue Leakage :** Tingginya angka *churn* menyebabkan kebocoran pendapatan bulanan/tahunan yang menghambat pertumbuhan perusahaan.
* **Ketidakefektifan Strategi Retensi :** Tanpa wawasan berbasis data, tim pemasaran sering kali melakukan promosi retensi untuk mempertahankan pelanggan secara acak, yang mana hal tersebut tidak efisien dan membuang anggaran pada pelanggan yang sebenarnya setia atau sudah pasti akan pergi (*churn*).

### 1.6 Metrik Evaluasi

Dalam konteks *churn prediction*, tingkat akurasi (*Accuracy*) seringkali bias karena jumlah pelanggan yang setia biasanya jauh lebih banyak daripada yang *churn* (imbalanced data). Fokus pada proyek ini adalah mendeteksi pelanggan yang *churn* karena kita ingin mengurangi kerugian atau loss profit yang disebabkan oleh pelanggan *churn*, sehingga kelas positif (1) diberikan kepada pelanggan yang *churn* dan kelas negatif (0) diberikan kepada pelanggan yang tetap setia berlangganan. Pemilihan metrik didasarkan pada **Cost of Error** yang ditanggung oleh Stakeholder, dengan penekanan bahwa kesalahan False Negative memiliki dampak bisnis yang jauh lebih besar dibandingkan False Positive.

* **False Negative (FN):** Model memprediksi pelanggan "Setia", padahal aslinya "Churn".
    * Perusahaan kehilangan pelanggan berharga tanpa sempat melakukan pencegahan. Ini adalah kerugian finansial terbesar (*High Cost*) berupa hilangnya *Customer Lifetime Value* (CLV).
* **False Positive (FP):** Model memprediksi pelanggan "Churn", padahal aslinya "Setia".
    * Perusahaan memberikan promosi retensi kepada pelanggan yang sebenarnya tidak berniat pergi. Biayanya relatif kecil, hanya sebatas biaya promosi/diskon (*Low Cost*).

**Kesimpulan Pemilihan Metrik:**
Berdasarkan pertimbangan biaya di atas, prioritas Stakeholder adalah meminimalisir kehilangan pelanggan (FN). Maka, metrik utama yang digunakan adalah:

1. **PR AUC (Precision–Recall Area Under Curve) :** Digunakan sebagai metrik utama karena mampu mengevaluasi trade-off antara precision dan recall secara menyeluruh pada berbagai nilai threshold, serta lebih representatif untuk imbalanced data
2. **Recall (Sensitivity):** Fokus utama untuk menangkap sebanyak mungkin pelanggan yang berpotensi *churn*. Semakin tinggi *Recall*, semakin sedikit pelanggan *churn* yang lolos dari deteksi.
3.  **Precision:** Untuk mengontrol tingkat kesalahan False Positive agar biaya promosi tetap efisien.


### 1.7 *5-Point Business ML Goals*

1.  **Problem :**
    Tingginya tingkat *churn* yang tidak terdeteksi secara dini menyebabkan hilangnya pendapatan berulang (*recurring revenue*) dan inefisiensi biaya akuisisi.
2.  **Data :**
    Menggunakan data historis perusahaan yang mencakup demografi pelanggan dan riwayat transaksi.
3.  **ML Objective :**
    Membangun model klasifikasi biner untuk memprediksi probabilitas seorang pelanggan akan berhenti berlangganan (*churn probability*) di bulan berikutnya.
4.  **Action :**
    Mengintegrasikan hasil prediksi kepada tim bisnis/marketing, di mana pelanggan berisiko tinggi untuk *churn* akan diberikan treatment khusus berupa insentif personal, diskon, atau pendekatan *customer service* proaktif.
5.  **Value :**
    Menurunkan *Churn Rate*, menyelamatkan pendapatan potensial (*Saved Revenue*), dan meningkatkan efisiensi biaya pemasaran dengan menargetkan retensi hanya pada pelanggan yang benar-benar berisiko.

## 5. Data Preparation

### Data Splitting

* X merupakan fitur untuk mendeteksi target/label y
* X semua variabel selain *Churn*
* y adalah target yaitu `Churn`
* Data ada di *split* menggunakan metode stratify agar tidak data train dan split tidak tercampur serta proporsi antar data nya sama atau seimbang

```python
X = df.drop(columns='Churn')
y = df['Churn'].map({'Yes': 1, 'No': 0})

# default value test_size = 0.25 kalo train_size dan test_size kosong
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify = y, random_state = RANDOM_STATE)
print("Train:", X_train.shape, "Test:", X_test.shape)
```

### Feature Engineering

```python
cat_features=[ 'OnlineSecurity', 'OnlineBackup',
       'InternetService', 'DeviceProtection', 'TechSupport', 'Contract']
binary_features=['Dependents','PaperlessBilling']
num_features=['tenure', 'MonthlyCharges']

num_pipeline = Pipeline(steps=[
('imputer', SimpleImputer(strategy='median')),
('scaler', StandardScaler())
])
binary_pipeline = Pipeline(steps=[
('imputer', SimpleImputer(strategy='most_frequent')),
('encoder', OneHotEncoder(drop='if_binary', handle_unknown='ignore', sparse_output=False))
])
cat_pipeline = Pipeline(steps=[
('imputer', SimpleImputer(strategy='most_frequent')),
('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])


preprocessor = ColumnTransformer(
transformers=[
('num', num_pipeline, num_features),
('binary', binary_pipeline, binary_features),
('cat', cat_pipeline, cat_features)
],
remainder='drop'
)
```

```python
testing = pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
testing.head()
```

### Model Benchmarking : K-Fold

```python
def benchmark_model(prepocessor, name, model, X_train, y_train):
    estimator = Pipeline([
        ('preprocess', preprocessor),
        ('model', model)])

    cv = StratifiedKFold(
        n_splits=5    )

    start = time.time()

    cv_results = cross_validate(
        estimator,
        X_train,
        y_train,
        cv=cv,
        scoring='average_precision',
        n_jobs=-1
    )

    elapsed = time.time() - start

    mean_pr_auc = cv_results['test_score'].mean()
    std_pr_auc = cv_results['test_score'].std()

    return {
        'Model': name,
        'PR_AUC (CV)': mean_pr_auc,
        'PR_AUC_STD': std_pr_auc,
        'Training Time (s)': elapsed
    }
```

```python
results = []
for name, model in models:
    print(f"Benchmarking {name}...")
    res = benchmark_model(preprocessor, name, model, X_train, y_train)
    results.append(res)


benchmark_df = pd.DataFrame(results).sort_values(by='PR_AUC (CV)', ascending=False)
benchmark_df.reset_index(drop=True, inplace=True)


benchmark_df
```

### Resampling pada Model Logistic Regression

Model **Logistic Regression** menggunakan teknik Resampling untuk menangani imbalanced data.
Pada proses ini diuji antara Random Over Sampling, Random Under Sampling dan SMOTE untuk melihat teknik resampling mana yang paling bagus digabungkan dengan Logistic Regression pada kasus ini.

```python
def benchmark_pipeline(name, pipeline, X_train, y_train):
    cv = StratifiedKFold(
        n_splits=5,
        shuffle=True,
        random_state=42
    )

    start = time.time()

    cv_results = cross_validate(
        pipeline,
        X_train,
        y_train,
        cv=cv,
        scoring='average_precision',
        n_jobs=-1
    )

    elapsed = time.time() - start

    return {
        'Model': name,
        'PR_AUC (CV)': cv_results['test_score'].mean(),
        'PR_AUC_STD': cv_results['test_score'].std(),
        'Training Time (s)': elapsed
    }
```

```python
logreg = LogisticRegression(
    random_state=RANDOM_STATE,
    max_iter=1000
)

pipelines = {
    'Baseline (No Resampling)': Pipeline([
        ('preprocess', preprocessor),
        ('model', logreg)
    ]),
    'Random Over Sampling': ImbPipeline([
        ('preprocess', preprocessor),
        ('sampler', RandomOverSampler(random_state=RANDOM_STATE)),
        ('model', logreg)
    ]),
    'Random Under Sampling': ImbPipeline([
        ('preprocess', preprocessor),
        ('sampler', RandomUnderSampler(random_state=RANDOM_STATE)),
        ('model', logreg)
    ]),
    'SMOTE': ImbPipeline([
        ('preprocess', preprocessor),
        ('sampler', SMOTE(random_state=RANDOM_STATE)),
        ('model', logreg)
    ])
}
```

```python
results = []

for name, pipe in pipelines.items():
    results.append(
        benchmark_pipeline(name, pipe, X_train, y_train)
    )

resampling_df = pd.DataFrame(results).sort_values(by='PR_AUC (CV)', ascending=False)
resampling_df.reset_index(drop=True, inplace=True)
resampling_df
```

### Penalized pada Model Logistic Regression

```python
logreg_penalized = LogisticRegression(
    random_state=42,
    max_iter=1000,
    class_weight='balanced'
)
```

```python
penalized_pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('model', logreg_penalized)
])

start = time.time()

cv_results = cross_validate(
    penalized_pipeline,
    X_train,
    y_train,
    cv=StratifiedKFold(n_splits=5),
    scoring='average_precision',
    n_jobs=-1
)

elapsed = time.time() - start

penalized_result = pd.DataFrame([{
    'Model': 'Logistic Regression (Penalized)',
    'PR_AUC (CV)': cv_results['test_score'].mean(),
    'PR_AUC_STD': cv_results['test_score'].std(),
    'Training Time (s)': elapsed
}])

penalized_result
```

```python
final_comparison_df = pd.concat(
    [resampling_df, penalized_result],
    ignore_index=True
)

final_comparison_df.sort_values(by='PR_AUC (CV)', ascending=False).reset_index(drop=True)
```

```python
random_search = RandomizedSearchCV(
    estimator=pipe_rus_logreg,
    param_distributions=param_dist,
    n_iter=20,
    scoring='average_precision',   # PR AUC
    cv=StratifiedKFold(n_splits=5),
    n_jobs=-1,
    verbose=1,
    random_state=42)

random_search.fit(X_train, y_train)
```
